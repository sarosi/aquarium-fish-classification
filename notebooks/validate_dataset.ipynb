```python
import os
from PIL import Image

DATASET_PATH = "../dataset"
REMOVE_BAD = True  # set to True to auto-delete corrupted files

valid_ext = ('.jpg', '.jpeg', '.png')

total_images = 0
bad_files = []
class_counts = {}

for split in ["train", "val"]:
    split_path = os.path.join(DATASET_PATH, split)
    
    if not os.path.exists(split_path):
        print(f"[WARNING] Missing split: {split}")
        continue

    print(f"\n--- Checking {split.upper()} ---")

    classes = [d for d in os.listdir(split_path)
               if os.path.isdir(os.path.join(split_path, d))]

    for cls in classes:
        cls_path = os.path.join(split_path, cls)
        count = 0

        for file in os.listdir(cls_path):
            if file.startswith("._") or file == ".DS_Store":
                continue

            if not file.lower().endswith(valid_ext):
                continue

            path = os.path.join(cls_path, file)

            try:
                img = Image.open(path)
                img.verify()
                count += 1
                total_images += 1
            except Exception:
                bad_files.append(path)

        class_counts[f"{split}/{cls}"] = count

# Print results
print("\n=== DATASET SUMMARY ===")
for k, v in class_counts.items():
    print(f"{k}: {v} images")

print(f"\nTotal valid images: {total_images}")
print(f"Bad images found: {len(bad_files)}")

# Show bad files
if bad_files:
    print("\n--- BAD FILES ---")
    for f in bad_files[:20]:
        print(f)
    if len(bad_files) > 20:
        print(f"... and {len(bad_files) - 20} more")

# Optional removal
if REMOVE_BAD and bad_files:
    for f in bad_files:
        os.remove(f)
    print("\nBad files removed.")

# Check imbalance
print("\n=== CLASS BALANCE CHECK ===")
train_counts = [v for k, v in class_counts.items() if k.startswith("train/")]

if train_counts:
    min_c = min(train_counts)
    max_c = max(train_counts)
    ratio = max_c / min_c if min_c > 0 else 0

    print(f"Min: {min_c}, Max: {max_c}, Ratio: {ratio:.2f}")

    if ratio > 2:
        print("⚠️ Warning: Class imbalance detected (ratio > 2)")
    else:
        print("✅ Class balance looks reasonable")
```
